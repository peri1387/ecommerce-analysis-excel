## Advanced Analysis: Order Volume vs Revenue Validation



This additional analysis validates the main findings of the dashboard.
**Scatter plot** and **linear regression** confirm a near-linear relationship between daily orders and revenue (*correlation: 0.96, R² ≈ 0.92*), indicating that revenue performance is primarily driven by order volume.
**Average Order Value** remains stable throughout the month, reinforcing the volume-driven revenue model.



## Key Insights

* **Revenue concentration**
  Electronics (31%) and Fashion (30%) together generate over 60% of total revenue, indicating strong reliance on two core categories.

* **Platform balance**
  Revenue is evenly distributed across platforms, suggesting low channel dependency and reduced platform risk.

* **Seasonality**
  Monthly revenue remains relatively stable throughout the year, with no strong seasonal spikes, indicating consistent demand.

* **Ratings vs Reviews**
  Higher review volume does not necessarily translate to higher average ratings, suggesting that popularity does not always reflect customer satisfaction.

* **Brand performance**
  Brand-level sales volumes are relatively balanced, indicating no single brand dominance in unit sales.

* **Correlation and regression analysis** further validate that revenue fluctuations are almost entirely explained by changes in order volume, while Average Order Value remains stable.


## Business Recommendations

- **Strengthen core categories (Electronics & Fashion):**  
  Since over 60% of total revenue comes from these two categories, the business should prioritize inventory optimization, targeted promotions, and product expansion in these segments while closely monitoring dependency risk.

- **Maintain multi-platform presence:**  
  The balanced revenue distribution across platforms suggests a healthy channel mix. The company should continue investing evenly across platforms to reduce dependency risk and maintain negotiation power with partners.

- **Leverage stable demand for planning:**  
  With no strong seasonal revenue spikes, demand forecasting and inventory planning can follow a steady, year-round strategy, reducing the need for aggressive seasonal stock adjustments.

- **Focus on quality over volume in customer feedback:**  
  Since higher review volumes do not guarantee higher ratings, efforts should focus on improving product quality and post-purchase experience rather than solely driving review counts.

- **Develop brand differentiation strategies:**  
  Balanced brand-level sales indicate strong competition. The company can create differentiation through exclusive partnerships, bundled offers, or brand-focused marketing campaigns to drive preference and loyalty.


## Final Notes

This analysis was conducted using Excel pivot tables an pivot charts, KPIs, and an interactive dashboard.
Advanced correlation and regression analysis was added to validate the relationship between order volume and revenue.
The project demonstrates how business insights can be extracted and communicated effectively without advanced modeling techniques.
